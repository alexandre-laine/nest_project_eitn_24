```python
from utils.capacity import compute_capacity_sequential
from utils.inputs import generate_piecewise_constant_signal
from utils.extractors import filter_spikes_parallel
import pandas as pd
```

```python
step_duration = 20.
num_steps = 200

sim_time = num_steps*step_duration
filter_tau = 20. # [ms]

nEnc_range = np.linspace(0, 1_000, 51, dtype=int)[1:]

NE = 1000
```

```python
sig, times, inp_times, inp_amplitudes = generate_piecewise_constant_signal(
        seed=1, 
        num_steps=num_steps,
        step_duration=step_duration, 
        resolution=dt, 
        scale=scale
)
```

```python
enc_activity
```

```python
main_capacity = [np.load(os.path.join(folder_name, "num_encneuron", f"main_capacity_numencneuron_{n_Enc}.npy")) for n_Enc in nEnc_range]
```

```python
Capacities = []

for n_Enc in nEnc_range:

    main_activity = pd.read_csv(os.path.join(folder_name, "num_encneuron", f"main_activity_numencneuron_{n_Enc}.csv"))

    main_states = filter_spikes_parallel(
        main_activity['times'].to_numpy(), main_activity['senders'].to_numpy(), 
        NE, 
        t_start=0., t_stop=sim_time, dt=dt, 
        tau=filter_tau, 
        n_processes=16
    )

    time_capacity, encoder_capacity, main_capacity = compute_capacity_sequential(
        time_vector=times,
        signal=sig[1:], 
        enc_states=enc_states,
        ex_states=main_states,
        subsampling_factor=subsampling_factor,
        max_lag=max_lag,
        step_lag=step_lag,
        dt=dt
    )

    Capacities.append(main_capacity)
```

```python
Capacities = np.array(Capacities)
print(Capacities.shape)
```
